```python
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt

# Load pretrained model
model = ResNet50(weights='imagenet')

# Correct image path
img_path = '/kaggle/working/dog.jpg'

# Load and preprocess
img = image.load_img(img_path, target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)

# Predict
preds = model.predict(x)

# Decode top predictions
print("‚úÖ Top Predictions:")
for i, (imagenet_id, label, score) in enumerate(decode_predictions(preds, top=3)[0]):
    print(f"{i+1}. {label}: {score*100:.2f}%")

# Show image with prediction
plt.imshow(img)
plt.axis('off')
plt.title("Predicted Label: " + decode_predictions(preds, top=1)[0][0][1])
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.preprocessing.image import img_to_array, load_img
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.models import Model

# Step 1Ô∏è‚É£ ‚Äî Convert .avif to .jpg
input_path = '/kaggle/input/yyyyyyyyy/most-popular-dog-breeds-bracco-italiano-68096ed33c726.avif'
output_path = '/kaggle/working/dog.jpg'

# Read using OpenCV
img_bgr = cv2.imread(input_path)

if img_bgr is None:
    print("‚ùå Could not read the AVIF file. Please check the folder name (yyyyyyyyy).")
else:
    # Convert color (BGR ‚Üí RGB)
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    # Save as JPG for Keras compatibility
    cv2.imwrite(output_path, cv2.cvtColor(img_rgb, cv2.COLOR_RGB2BGR))
    print("‚úÖ Converted image saved as:", output_path)

    # Step 2Ô∏è‚É£ ‚Äî Load pretrained ResNet50
    base_model = ResNet50(weights='imagenet')

    # Step 3Ô∏è‚É£ ‚Äî Extract outputs from the first convolutional layer
    model = Model(inputs=base_model.inputs, outputs=base_model.layers[1].output)

    # Step 4Ô∏è‚É£ ‚Äî Load the converted JPG image
    image = load_img(output_path, target_size=(224, 224))
    image_array = img_to_array(image)
    image_array = np.expand_dims(image_array, axis=0)
    image_array = preprocess_input(image_array)

    # Step 5Ô∏è‚É£ ‚Äî Get feature maps
    feature_maps = model.predict(image_array)
    print("‚úÖ Feature map shape:", feature_maps.shape)

    # Step 6Ô∏è‚É£ ‚Äî Display first 9 feature maps
    square = 3
    fig, axes = plt.subplots(square, square, figsize=(10, 10))
    for i in range(square * square):
        ax = axes[i // square, i % square]
        ax.imshow(feature_maps[0, :, :, i], cmap='gray')
        ax.axis('off')
    plt.suptitle("Feature Maps from First Layer")
    plt.show()
```

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications.resnet50 import preprocess_input
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Step 1Ô∏è‚É£ ‚Äî Load your ResNet50 model
model2 = ResNet50(weights='imagenet')

# Step 2Ô∏è‚É£ ‚Äî Choose which layers you want feature maps from
layers_index = [2, 5, 9, 13, 17]
outputs = [model2.layers[i].output for i in layers_index]

# Step 3Ô∏è‚É£ ‚Äî Create a new model that gives multiple outputs
model3 = Model(inputs=model2.inputs, outputs=outputs)
print("‚úÖ Model with multiple layer outputs created!")

# Step 4Ô∏è‚É£ ‚Äî Load your image
img_path = '/kaggle/input/yyyyyyyyy/most-popular-dog-breeds-bracco-italiano-68096ed33c726.avif'
img_bgr = cv2.imread(img_path)

if img_bgr is None:
    print("‚ùå Could not read AVIF file. Please check your dataset name (yyyyyyyyy).")
else:
    # Convert to RGB and save as JPEG (Keras can't load .avif)
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    cv2.imwrite('/kaggle/working/dog.jpg', cv2.cvtColor(img_rgb, cv2.COLOR_RGB2BGR))
    print("‚úÖ Converted AVIF image to JPG successfully!")

    # Step 5Ô∏è‚É£ ‚Äî Preprocess image for model input
    img = load_img('/kaggle/working/dog.jpg', target_size=(224, 224))
    x = img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)

    # Step 6Ô∏è‚É£ ‚Äî Generate feature maps from selected layers
    feature_maps = model3.predict(x, verbose=0)
    print("‚úÖ Feature maps extracted for layers:", layers_index)

    # Step 7Ô∏è‚É£ ‚Äî Visualize some feature maps from each layer
    for i, fmap in enumerate(feature_maps):
        print(f"üîπ Layer {layers_index[i]} shape: {fmap.shape}")

        # Show first 6 feature maps from this layer
        n_filters = 6
        plt.figure(figsize=(12, 6))
        plt.suptitle(f"Feature Maps from Layer {layers_index[i]}", fontsize=14)
        for j in range(n_filters):
            plt.subplot(2, 3, j + 1)
            plt.imshow(fmap[0, :, :, j], cmap='gray')
            plt.axis('off')
            plt.title(f'Filter {j+1}')
        plt.tight_layout()
        plt.show()
```
